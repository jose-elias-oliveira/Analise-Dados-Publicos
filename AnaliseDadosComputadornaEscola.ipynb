## A PRESENSÇA DE COMPUTADORES NO ÂMBITO ESCOLAR ##

Analisar os dados publicos do Censo Escolar 2018, 2016 para criar gráficos que represeta o quanto computadores estão conectados com o sistema de ensino por todo o brasil.

```python

import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

```

```python
dados2018 = pd.read_csv("ESCOLAS.csv", delimiter='|', usecols=['CO_ENTIDADE','NO_ENTIDADE', 'CO_REGIAO','CO_UF', 'CO_MUNICIPIO', 'IN_COMPUTADOR', 'QT_COMP_ALUNO'], encoding='iso-8859-1')
```

```python
dados2017 = pd.read_csv("ESCOLAS2017.csv",  delimiter='|',  usecols=['CO_ENTIDADE','NO_ENTIDADE', 'CO_REGIAO','CO_UF', 'CO_MUNICIPIO', 'IN_COMPUTADOR'], encoding='iso-8859-1')
```

```python
dados2016 = pd.read_csv("ESCOLAS2016.csv", delimiter='|',  usecols=['CO_ENTIDADE','NO_ENTIDADE', 'CO_REGIAO','CO_UF', 'CO_MUNICIPIO', 'IN_COMPUTADOR'], encoding='iso-8859-1')
```

### DADOS ESCOLAS POR REGIÃO (2018)


##### Regiões do País por quantidade de escolas com computadores (2018).


---

Trocando 'CO_REGIAO' pelo nome da região e 'CO_UF' pela sigla do estado


```python
de_para_regiao = {1:'NORTE',2:'NORDESTE',3:'SUDESTE',4:'SUL',5:'CENTRO-OESTE'} 
de_para_estado ={11:'RO',
12:'AC',
13:'AM',
14:'RR',
15:'PA',
16:'AP',
17:'TO',
21:'MA',
22:'PI',
23:'CE',
24:'RN',
25:'PB',
26:'PE',
27:'AL',
28:'SE',
29:'BA',
31:'MG',
32:'ES',
33:'RJ',
35:'SP',
41:'PR',
42:'SC',
43:'RS',
50:'MS',
51:'MT',
52:'GO', 
53:'DF'}
```

---

##### Regiões do País por quantidade de escolas com computadores (2018).

```python

dados2018['REGIAO'] = dados2018['CO_REGIAO'].apply(lambda x: de_para_regiao[x])
dados2018['ESTADO'] = dados2018['CO_UF'].apply(lambda x: de_para_estado[x])
dados2018.query('IN_COMPUTADOR == 1')['REGIAO'].value_counts().sort_values()
```

#### Estados brasileiros por quantidade de escolas com computadores (2018)

```python
dados2018.query('IN_COMPUTADOR == 1')['ESTADO'].value_counts().head(26).plot(kind='bar', figsize=(7,7))
```

### Gráfico quantidade de escolas que possuem computadores por região (2018)

```python
dados2018.query('IN_COMPUTADOR == 1')['REGIAO'].value_counts().head().plot(kind='pie', figsize=(10,10))
```

#### Quantidade de computadores por aluno (2018)

Quantidades de computadores disponiveis para uso dos alunos, mostra que muitas escolas no brasil ainda não possuem computadores para uso de seus alunos.

```python
dados2018.groupby('QT_COMP_ALUNO').size().sort_values().tail(20).plot(kind='barh', figsize=(7,7))
```
